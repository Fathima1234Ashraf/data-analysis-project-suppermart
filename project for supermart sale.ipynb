# FIND THE ANALYSIS QUESTIONS

What are the top 5 cities with the highest total sales?python code with diagram

```python
import pandas as pd
import matplotlib.pyplot as plt  # Correct import for plotting

# Calculate top 5 cities by sales
top_cities_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5)

# Plot the bar chart
plt.figure(figsize=(10, 6))
top_cities_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 5 Cities with Highest Total Sales', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

Which product category generates the most profit

```python
import pandas as pd

# Group data by Category and calculate the total profit for each
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

# Display the product category with the highest profit
most_profitable_category = category_profit.idxmax()
highest_profit = category_profit.max()

print(f"The product category generating the most profit is '{most_profitable_category}' with a total profit of {highest_profit:.2f}.")
```

How does profit vary across different regions

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data by Region and calculate total profit for each
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

# Plot the data
plt.figure(figsize=(10, 6))
region_profit.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Profit Variation Across Regions', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Profit', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

What is the monthly trend of total sales

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure the 'Order Date' column is in datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract year and month from the order date
df['Year-Month'] = df['Order Date'].dt.to_period('M')

# Group by 'Year-Month' and calculate total sales
monthly_sales = df.groupby('Year-Month')['Sales'].sum()

# Plot the monthly sales trend
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o', color='blue')
plt.title('Monthly Trend of Total Sales', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

Which state gives the highest average discount

```python
import pandas as pd

# Group data by State and calculate the average discount for each state
state_avg_discount = df.groupby('State')['Discount'].mean().sort_values(ascending=False)

# Display the state with the highest average discount
highest_discount_state = state_avg_discount.idxmax()
highest_discount_value = state_avg_discount.max()

print(f"The state with the highest average discount is '{highest_discount_state}' with an average discount of {highest_discount_value:.2f}%.")
```

What is the correlation between sales, discount, and profit

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix for sales, discount, and profit
correlation_matrix = df[['Sales', 'Discount', 'Profit']].corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Between Sales, Discount, and Profit', fontsize=16)
plt.show()
```

How do sales and profit distribute across different sub-categories

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame loaded with the Supermart Grocery dataset
# Load your data (ensure you've already read the dataset into 'df')
# df = pd.read_csv("path_to_your_file.csv")  # Uncomment and update the file path if needed

# Group data by 'Sub Category' and calculate total sales and profit for each sub-category
sub_category_sales_profit = df.groupby('Sub Category')[['Sales', 'Profit']].sum()

# Plotting the distribution of sales and profit across sub-categories
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Sales bar chart
sub_category_sales_profit['Sales'].plot(kind='bar', color='skyblue', edgecolor='black', ax=ax[0])
ax[0].set_title('Sales Distribution Across Sub-Categories', fontsize=16)
ax[0].set_xlabel('Sub-Category', fontsize=12)
ax[0].set_ylabel('Total Sales', fontsize=12)
ax[0].tick_params(axis='x', rotation=45)
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

# Profit bar chart
sub_category_sales_profit['Profit'].plot(kind='bar', color='lightgreen', edgecolor='black', ax=ax[1])
ax[1].set_title('Profit Distribution Across Sub-Categories', fontsize=16)
ax[1].set_xlabel('Sub-Category', fontsize=12)
ax[1].set_ylabel('Total Profit', fontsize=12)
ax[1].tick_params(axis='x', rotation=45)
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

What percentage of total sales comes from each region

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data by Region and calculate total sales for each region
region_sales = df.groupby('Region')['Sales'].sum()

# Calculate total sales across all regions
total_sales = region_sales.sum()

# Calculate the percentage of total sales from each region
region_sales_percentage = (region_sales / total_sales) * 100

# Plot the percentage sales by region
plt.figure(figsize=(8, 8))
region_sales_percentage.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'salmon', 'gold'], 
                             startangle=90, legend=False, figsize=(8, 8))

plt.title('Percentage of Total Sales from Each Region', fontsize=16)
plt.ylabel('')  # Remove the ylabel to clean up the chart
plt.show()
```

What is the relationship between discount and profit

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the correlation between Discount and Profit
correlation = df['Discount'].corr(df['Profit'])

# Display the correlation value
print(f"Correlation between Discount and Profit: {correlation}")

# Create a scatter plot to visualize the relationship between Discount and Profit
plt.figure(figsize=(8, 6))
plt.scatter(df['Discount'], df['Profit'], alpha=0.5, color='blue')
plt.title('Relationship Between Discount and Profit', fontsize=16)
plt.xlabel('Discount', fontsize=12)
plt.ylabel('Profit', fontsize=12)
plt.grid(True)
plt.show()
```

Which customer has generated the most revenue

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("D:\\datascience\\dataset\\Supermart Grocery Sales - Retail Analytics Dataset (2).csv")

# Check the column names to ensure correct reference
print("Column names:", df.columns)

# Strip any leading/trailing spaces from column names to avoid KeyErrors
df.columns = df.columns.str.strip()

# Now, check again for correct column names
print("After stripping spaces, column names:", df.columns)

# Group data by 'Customer Name' and calculate total sales for each customer
customer_sales = df.groupby('Customer Name')['Sales'].sum()

# Find the customer with the highest total sales
top_customer = customer_sales.idxmax()  # Customer Name with the highest sales
top_customer_sales = customer_sales.max()  # Total sales of that customer

# Display the result
print(f"Customer '{top_customer}' has generated the most revenue with total sales of {top_customer_sales}.")
```
